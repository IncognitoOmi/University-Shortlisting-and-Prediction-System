```python
import numpy as mmp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
df = df.loc[:,'GRE Score':]
```

```python
df1 = df.rename({'GRE Score':'GRE_SCORE','TOEFL Score':'TOEFL_SCORE','University Rating':'UNIVERSITY_RATING'}, axis=1)
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1.describe()
```

```python
plt.(df1['GRE_SCORE'], edgecolor='black', bins=20)
plt.show()
```

```python
Y = df1['Admit']
X = df1.drop(['Admit'], axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test ,y_train, y_test = train_test_split(X,Y,test_size=0.3)
```

```python
X_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
predicted = lr.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error, r2_score
```

```python
print(f'RMSE : {root_mean_squared_error(y_test, predicted)}')
print(f'R2 : {r2_score(y_test, predicted)}')
```
